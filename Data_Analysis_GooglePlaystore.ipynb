### Total number of Free and Paid Apps

```python
f=0
p=0
for i in df["Type"]:
    if(i=="Free"):
        f+=1
    else:
        p+=1
print("Free apps:",f,", Paid Apps:",p,)
```

```python
fp=f/t * 100
round(fp,2)
```

```python
pp=p/t * 100
round(pp,2)
```

### Total Number of apps in each category

```python
categories = {}


for name in df["Category"].unique():
    cnt=0
    for i in df["Category"]:
        if(i==name):
            cnt+=1
        categories[name]=cnt

print(categories)
```

```python
type(categories)
```

### Total Number of apps in each Type

```python
types = {}
for name in df["Type"].unique():
    cnt=0
    for i in df["Type"]:
        if(i==name):
            cnt+=1
            types[name]=cnt
print(types)
```

###  Total Number of apps in each Content rating

```python
c_ratings = {}
for name in df["Content Rating"].unique():
    cnt=0
    for i in df["Content Rating"]:
        if(i==name):
            cnt+=1
            c_ratings[name]=cnt
print(c_ratings)
```
